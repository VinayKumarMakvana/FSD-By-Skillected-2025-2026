# **Argument and Key-Argument**

```python
# Advance Function :- *args & **kwargs
# *args :- its allows a function to accept any number of positional arguments. this is tuppel

def show(*args):
  print(args)
show(10,20,30)
```

```python
def add(*args):
  total=0
  for num in args:
    total+=num
  return total
print(add(10,20))
print(add(10,20,30,50))
```

```python
def show(*args):
  for i in args:
    print(i)
```

```python
show(10,20,30)
```

```python
def names(*args):
  print("student are: ")
  for names in args:
    print(names)
names("vinay","akash","payal","tannu")
```

```python
def show(*number):
  print(type(number))
show(1,2,3)
```

```python
# **kwargs :- **kwargs allow a function to accept multiples key values input. this is dictionary
def show(**kwargs):
  print(kwargs)
```

```python
show(name="vinay",age=19)
```

```python
def student(**kwargs):
  for key,value in kwargs.items():
    print(key,":",value)
student(name="Vinay",age=19,course="BCA")
```

```python
def full_data(*args,**kwargs):
  print("Marks:",args)
  print("details:")
  for key,value in kwargs.items():
    print(key,":",value)
full_data(80,85,90,name="vinay",age=19)
```
