```python
# First file
# digita notebook

file=open("demo.text","w") # open() -> he is open your file [w-> wirte mode]
file.write("Hello World")
file.close()
# file.write("hello student") # already colse the file for this reason given erorr
```

```python
file=open("demo.text","r") #r(read mode)
data=file.read()# all content read
print(data) #screen output
file.close()
```

```python
file=open("demo.text","a") # append mode (old data safe)
file.write("\nWelcom to python")
file.close()
```
