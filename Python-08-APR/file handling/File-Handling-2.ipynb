# **File Handling-2**

```python
# Cursor -> Cursor is a current position inside a File.
with open("student.txt","w") as file:
  file.write("Vinay\nShubham")
```

```python
# tell() :- tell() [function] current cursor positon
# seek() :- seek() change cursor position
```

```python
with open("student.txt","r") as file:
  print(file.tell())
```

```python
with open("student.txt","r") as file:
  file.read(2)
  print(file.tell())
```

```python
# seek() :- moves cursir ti a specific position
with open("student.txt","r") as file:
  file.seek(3)
  print(file.read())
```

```python
# r+
with open("student.txt","r+") as file:
  data=file.read()
  print(data)
  file.write("\nNeha")
```

```python
with open("student.txt","r+") as file:
  data=file.read()
  print(data)
```

```python
with open("student.txt","r+") as file:
  file.write("admin")
```

```python
with open("student.txt","r+") as file:
  data=file.read()
  print(data)
```

```python
# What is w+ ?
with open("student.txt","w+") as file:
  file.write("Priya")
  print(file.read())
  file.seek(0)
  print(file.read(3))
  print(file.read())
```
