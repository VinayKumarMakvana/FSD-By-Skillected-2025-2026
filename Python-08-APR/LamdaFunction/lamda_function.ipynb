```python
#function-reuse
#module-multiple function ko organzize
#package -multiple modules organize
def add(a,b):
  return a+b

print(add(10,20))
```

```python
#need
#tempory function  lamda function:-a lamda function is a small anonymous function written in a single line.
#is a shortcut way of creating a function
#lambda=one line function.
#anonymous -without name
```

```python
#normal function
def square(x):
  return x*x
print(square(5))
```

```python
#lamda version :
square=lambda x:x*x
print(square(5))
```

```python
def add(a,b):
  return a+b
print(add(10,20))
```

```python
add=lambda a,b:a+b
print(add(10,20))
```

```python
bonus=lambda salary:salary+5000
print(bonus(30000))
```

```python
check=lambda age:"adult" if age>=18 else "minor"
print(check(20))
```

```python
marks=[10,20,30,40]
#map():-applies the same function to every element.
result=list(map(lambda x:x+5,marks))
print(result)
```

```python
#filter():-selects elements based on condition
ages=[15,18,22,16,25]
result=list(filter(lambda x:x%2==0,ages))
print(result)
```
