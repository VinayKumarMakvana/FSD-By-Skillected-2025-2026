```python
# Methods
# Q8. TV Class

class TV:
    def __init__(self):
        self.is_on = False
        self.volume = 10

    def power_on(self):
        self.is_on = True
        print("TV is now ON.")

    def power_off(self):
        self.is_on = False
        print("TV is now OFF.")

    def increase_volume(self):
        if self.is_on:
            self.volume += 1
            print(f"Volume increased to {self.volume}")
        else:
            print("TV is OFF. Turn it ON first.")

    def decrease_volume(self):
        if self.is_on and self.volume > 0:
            self.volume -= 1
            print(f"Volume decreased to {self.volume}")
        elif not self.is_on:
            print("TV is OFF. Turn it ON first.")


# Object create
tv = TV()

# Methods call
tv.power_on()
tv.increase_volume()
tv.increase_volume()
tv.decrease_volume()
tv.power_off()
```

```python
# Encapsulation
# Q19. Wallet Class (Private Variables)
class Wallet:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private variable

    def add_money(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Added ₹{amount}. New Balance: ₹{self.__balance}")

    def spend_money(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Spent ₹{amount}. Remaining Balance: ₹{self.__balance}")
        else:
            print("Insufficient funds or invalid amount!")

    def check_balance(self):
        return self.__balance

w = Wallet(500)
w.add_money(200)
w.spend_money(100)
```

```python
# Q22. Transport Abstract Class

class Transport(ABC):
    @abstractmethod
    def travel(self):
        pass

class Car(Transport):
    def travel(self):
        print("Traveling on highways by Car.")

class Bike(Transport):
    def travel(self):
        print("Splitting lanes and tracking narrow shortcuts by Bike.")

class Train(Transport):
    def travel(self):
        print("Long-distance mass transit commuting via Train tracks.")


# Object create
car = Car()
bike = Bike()
train = Train()

# Methods call
car.travel()
bike.travel()
train.travel()
```

```python
# Logic-Based OOPS Questions
# Q25. Student Result System

class StudentResultSystem:
    def __init__(self, student_name, marks_dict):
        # marks_dict expected structure: {"Math": 85, "Science": 90, ...}
        self.student_name = student_name
        self.marks = marks_dict

    def calculate_total(self):
        return sum(self.marks.values())

    def calculate_percentage(self):
        return (self.calculate_total() / (len(self.marks) * 100)) * 100

    def get_grade(self):
        pct = self.calculate_percentage()
        if pct >= 90: return "A+"
        elif pct >= 75: return "A"
        elif pct >= 60: return "B"
        elif pct >= 35: return "Pass"
        else: return "Fail"

    def display_report(self):
        print(f"\n--- Performance Card: {self.student_name} ---")
        for sub, score in self.marks.items():
            print(f"{sub}: {score}")
        print(f"Total Marks: {self.calculate_total()}")
        print(f"Percentage: {self.calculate_percentage()}%")
        print(f"Final Grade: {self.get_grade()}")

s1 = StudentResultSystem("Vinay", {"DSA": 88, "DBMS": 92, "Python": 95, "Web Dev": 90, "Networking": 85})
s1.display_report()
```

```python
# Q26. Shopping Cart
class ShoppingCart:
    def __init__(self):
        self.cart = {} # Key: Product Name, Value: Price

    def add_product(self, name, price):
        self.cart[name] = price
        print(f"Added {name} (₹{price}) to cart.")

    def remove_product(self, name):
        if name in self.cart:
            del self.cart[name]
            print(f"Removed {name} from cart.")
        else:
            print("Product item matching search string not found in cart.")

    def calculate_total(self):
        return sum(self.cart.values())

    def display_bill(self):
        print("\n=== Shopping Receipt ===")
        for item, price in self.cart.items():
            print(f"{item}: ₹{price}")
        print(f"Total Billable Value: ₹{self.calculate_total()}\n")

my_cart = ShoppingCart()
my_cart.add_product("Wireless Mouse", 800)
my_cart.add_product("HDMI Cable", 350)
my_cart.display_bill()
```

```python
# Q27. Movie Booking System

class MovieBookingSystem:
    def __init__(self, movie_name):
        self.movie_name = movie_name
        self.booked_seats = []

    def book_ticket(self, seat_num):
        if seat_num not in self.booked_seats:
            self.booked_seats.append(seat_num)
            print(f"Seat {seat_num} for movie '{self.movie_name}' successfully reserved.")
        else:
            print(f"Seat {seat_num} is already taken!")

    def cancel_ticket(self, seat_num):
        if seat_num in self.booked_seats:
            self.booked_seats.remove(seat_num)
            print(f"Booking cancelled for seat {seat_num}.")
        else:
            print("No booking found for this seat.")

    def display_details(self):
        print(f"Movie: {self.movie_name} | Booked Seats: {self.booked_seats}")


# Object create
movie = MovieBookingSystem("Pushpa 2")

# Methods call
movie.book_ticket(12)
movie.book_ticket(15)
movie.book_ticket(12)      # Already booked
movie.display_details()

movie.cancel_ticket(12)
movie.display_details()
```

```python
# Q28. Cab Booking System

class CabBookingSystem:
    def __init__(self, driver_name, cab_type):
        self.driver_name = driver_name
        self.cab_type = cab_type
        self.is_booked = False

    def book_cab(self):
        if not self.is_booked:
            self.is_booked = True
            print(f"Cab booking confirmed with {self.driver_name} ({self.cab_type}).")
        else:
            print("Cab is currently busy on an ongoing transit assignment.")

    def cancel_ride(self):
        if self.is_booked:
            self.is_booked = False
            print("Ride booking cancelled successfully.")
        else:
            print("No active booking to cancel.")

    def calculate_fare(self, distance_km):
        rate_per_km = 15 if self.cab_type == "Sedan" else 10
        return distance_km * rate_per_km


# Object create
cab = CabBookingSystem("Rahul", "Sedan")

# Methods call
cab.book_cab()
print("Fare:", cab.calculate_fare(20))
cab.cancel_ride()
```

```python
# Q29. Electricity Bill Generator

class ElectricityBillGenerator:
    def __init__(self, consumer_name, units_consumed):
        self.consumer_name = consumer_name
        self.units_consumed = units_consumed

    def calculate_bill(self):
        # Tiered step rates
        if self.units_consumed <= 100:
            return self.units_consumed * 5
        elif self.units_consumed <= 300:
            return (100 * 5) + ((self.units_consumed - 100) * 7)
        else:
            return (100 * 5) + (200 * 7) + ((self.units_consumed - 300) * 10)

    def print_bill(self):
        print("\n--- DISCOM Electricity Bill ---")
        print(f"Consumer Account Name: {self.consumer_name}")
        print(f"Energy Meter Consumption: {self.units_consumed} Units")
        print(f"Net Payable Amount: ₹{self.calculate_bill()}")


# Object create
bill = ElectricityBillGenerator("Rahul", 350)

# Method call
bill.print_bill()
```

# **Q30. Challenge: School Management System**

```python
from abc import ABC, abstractmethod

# Abstract Class
class School(ABC):
    @abstractmethod
    def school_info(self):
        pass


# Parent Class
class Person(School):
    def __init__(self, name, age):
        self.name = name
        self.age = age


# Child Class - Student
class Student(Person):
    def __init__(self, name, age, roll_no, fees):
        super().__init__(name, age)
        self.roll_no = roll_no
        self.__fees = fees      # Encapsulation (Private)

    def pay_fees(self, amount):
        self.__fees -= amount
        print(f"₹{amount} fees paid successfully.")

    def show_fees(self):
        print(f"Remaining Fees: ₹{self.__fees}")

    def school_info(self):
        print("Student Record")
        print(f"Name : {self.name}")
        print(f"Age  : {self.age}")
        print(f"Roll : {self.roll_no}")


# Child Class - Teacher
class Teacher(Person):
    def __init__(self, name, age, subject, salary):
        super().__init__(name, age)
        self.subject = subject
        self.__salary = salary      # Encapsulation (Private)

    def show_salary(self):
        print(f"Salary: ₹{self.__salary}")

    def school_info(self):
        print("Teacher Record")
        print(f"Name    : {self.name}")
        print(f"Age     : {self.age}")
        print(f"Subject : {self.subject}")


# Object Create
student = Student("Rahul", 20, 101, 50000)
teacher = Teacher("Sharma Sir", 40, "Python", 60000)

# Student Methods
student.school_info()
student.show_fees()
student.pay_fees(10000)
student.show_fees()

print()

# Teacher Methods
teacher.school_info()
teacher.show_salary()
```

# **Bonus Challenge (Interview Level): System Architecture Design**

```python
from abc import ABC, abstractmethod

# ------------------ Abstract ATM ------------------
class ATM(ABC):
    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def check_balance(self):
        pass


# ------------------ Bank Account ------------------
class BankAccount:
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.__balance = balance        # Encapsulation
        self.transactions = []

    def get_balance(self):
        return self.__balance

    def deposit_money(self, amount):
        self.__balance += amount
        self.transactions.append(f"Deposit ₹{amount}")

    def withdraw_money(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            self.transactions.append(f"Withdraw ₹{amount}")
            return True
        return False


# ------------------ ATM Machine ------------------
class ATMSystem(ATM):
    def __init__(self, account):
        self.account = account

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid deposit amount.")
            return

        self.account.deposit_money(amount)
        print(f"₹{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
            return

        if self.account.withdraw_money(amount):
            print(f"Please collect ₹{amount}.")
        else:
            print("Insufficient Balance.")

    def check_balance(self):
        print(f"Available Balance: ₹{self.account.get_balance()}")

    def mini_statement(self):
        print("\n----- Mini Statement -----")
        for transaction in self.account.transactions:
            print(transaction)


# ------------------ Main ------------------

account = BankAccount("Rahul", "1234567890", 10000)

atm = ATMSystem(account)

atm.check_balance()
atm.deposit(5000)
atm.withdraw(3000)
atm.withdraw(20000)
atm.check_balance()
atm.mini_statement()
```
