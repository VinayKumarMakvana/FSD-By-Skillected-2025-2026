## **Types of error :-**

```python
# 1). value error :- it's occurs when a function recive the data type but wrong value.
try:
  pin=int(input("enter your pin: "))
  print("login success")
except ValueError:
  print("pin must be contain number only")
```

```python
# 2). Zero Division Error
# it's occurs when a number is divided by zero.
try:
  num=int(input("enter a number: "))
  print(100/num)
except ZeroDivisionError as e:
  print("cannot divide by zero")
```

```python
# 3). Name Error :- Name error occurs when a variable is not defined.
try:
  print(age)
except NameError:
  print("Variable agee is not defined")
except:
  print("Something else went wrong")
```

```python
# 4). index error :- index error occurs when an individual index is accessed.
try:
  student=["akash","vinay"]
  print(student[1])
except IndexError:
  print("index out of range")
```

```python
# 5). File note found error :- occurs when the file dose not exist.
try:
  file=open("demo.txt","r")
  print(file.read())
except FileNotFoundError:
  print("file note Found")
```

```python
# 6). Multiple except blocks:-
try:
  numm= int(input("enter a number: "))
  print(100/numm)
except ValueError:
  print("invalid input please enter in number not in text")
except ZeroDivisionError:
  print("cannot divide by zero")
```

## **DAY-2**

```python
# finally block :- the finally block always executes whether an exception occurs or not.
# try:
   # risky code
# except:
   # error handling
# finally:
   # always run
```

```python
try:
  print("progam start")
except:
  print("Error")
finally:
  print("Program ended")
```

```python
try:
  print(10/0)
except:
  print("Error Found")
finally:
  print("Program ended")
```

```python
try:
  password=int(input("Enter your Password "))
  print("Login Successfull")
except ValueError:
  print("Password Show contain nummber")
finally:
  print("Session ended")
```

```python
try:
  file=open("student.txt","r")
  print(file.read())
except FileNotFoundError:
  print("File note Found")
finally:
  print("Closing File")
```

```python
# else block :- The else block execute only when no exception occuse in the try block.
# The else block is used to separate succsessful code from error handling.
# try:
   # risky code
# except:
   # error handling
# else:
    # sucsses code or message
# finally:
   # always run
```

```python
try:
  num=int(input("Enter Number "))
except ValueError:
  print("Only number allowed")
else:
  print("Input Succsessful")
```

```python
try:
  num=int(input("Enter number "))
  print(100/num)
except ValueError:
  print("Only numbers allowed")
except ZeroDivisionError:
  print("Can not Divide by Zero")
else:
  print("calculation successfull")
finally:
  print("Program ended")
```

```python
# as error :-
# the as key word stores the exception object into variable

try:
  num=int(input("Enter number: "))
except ValueError as e:
  print(e)
```
