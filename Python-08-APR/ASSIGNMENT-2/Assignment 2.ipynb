```python
# Q39. Create a function to check prime number
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num=int(input("Enter a prime number "))
print(is_prime(num))
```

```python
# Q43. Create a calculator using functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):

    if b == 0:
        return "Cannot divide by zero"

    return a / b


num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter choice: "))

if choice == 1:
    print("Result =", add(num1, num2))

elif choice == 2:
    print("Result =", subtract(num1, num2))

elif choice == 3:
    print("Result =", multiply(num1, num2))

elif choice == 4:
    print("Result =", divide(num1, num2))

else:
    print("Invalid Choice")
```

```python
# Q44. Function that takes a list and returns only even numbers
def even_numbers(numbers):

    even_list = []

    for num in numbers:
        if num % 2 == 0:
            even_list.append(num)

    return even_list


data = [1, 2, 3, 4, 5, 6, 7, 8]

print("Even Numbers:", even_numbers(data))
```

```python
# Q45. Function that finds second largest number in a list
def second_largest(numbers):

    numbers.sort()

    return numbers[-2]


data = [10, 40, 20, 90, 60]

print("Second Largest Number:", second_largest(data))
```

**PROJECT 1: STUDENT MANAGEMENT SYSTEM**

```python
students = []

def add_student():
    name = input("Enter name: ")
    age = int(input("Enter age: "))

    student = {"name": name, "age": age}
    students.append(student)
    print("Student added successfully!\n")

def show_students():
    if not students:
        print("No students found.\n")
        return

    print("\nStudent List:")
    for s in students:
        print(f"Name: {s['name']}, Age: {s['age']}")
    print()

# Main program
while True:
    print("1. Add Student")
    print("2. Show Students")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        show_students()
    elif choice == "3":
        break
    else:
        print("Invalid choice\n")
```

**PROJECT 2: SIMPLE BANK SYSTEM**

```python
balance = 0

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print("Deposited successfully!\n")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: "))

    if amount > balance:
        print("Insufficient balance!\n")
    else:
        balance -= amount
        print("Withdrawn successfully!\n")

def show_balance():
    print("Current balance:", balance, "\n")

# Main program
while True:
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Show Balance")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        deposit()
    elif choice == "2":
        withdraw()
    elif choice == "3":
        show_balance()
    elif choice == "4":
        break
    else:
        print("Invalid choice\n")
```

**PROJECT 3: NUMBER GUESSING GAME**

```python
import random

def play_game():
    number = random.randint(1, 100)
    attempts = 0

    print("Guess a number between 1 and 100")

    while True:
        guess = int(input("Enter your guess: "))
        attempts += 1

        if guess > number:
            print("Too high!")
        elif guess < number:
            print("Too low!")
        else:
            print(f"Correct! You guessed in {attempts} attempts.\n")
            break

# Run game
play_game()
```
