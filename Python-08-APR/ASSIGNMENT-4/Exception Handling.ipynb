```python
# 23. Marks Calculator Module

%%writefile marks.py

def total_marks(m1, m2, m3):
    return m1 + m2 + m3

def percentage(total):
    return total / 3

def grade(per):
    if per >= 90:
        return "A+"
    elif per >= 75:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 40:
        return "C"
    else:
        return "Fail"
```

```python
import marks

m1 = int(input("Enter Marks 1: "))
m2 = int(input("Enter Marks 2: "))
m3 = int(input("Enter Marks 3: "))

total = marks.total_marks(m1, m2, m3)
per = marks.percentage(total)
grade = marks.grade(per)

print("Total =", total)
print("Percentage =", per)
print("Grade =", grade)
```

```python
# 26. Banking System

balance = 5000

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Balance")
    print("4. Exit")

    try:
        choice = int(input("Enter Choice: "))

        if choice == 1:
            amount = int(input("Enter Deposit Amount: "))
            balance += amount
            print("Balance =", balance)

        elif choice == 2:
            amount = int(input("Enter Withdraw Amount: "))

            if amount > balance:
                raise Exception("Insufficient Balance")

            balance -= amount
            print("Balance =", balance)

        elif choice == 3:
            print("Current Balance =", balance)

        elif choice == 4:
            print("Thank You")
            break

        else:
            print("Invalid Choice")

    except ValueError:
        print("Enter numbers only")

    except Exception as e:
        print(e)
```

```python
# 28. Converter Module

%%writefile converter.py

def km_to_m(km):
    return km * 1000

def m_to_km(m):
    return m / 1000

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
```

```python
# Q30. Menu-Driven Application .

import os

# Package create
os.makedirs("college", exist_ok=True)

# __init__.py
with open("college/__init__.py", "w") as f:
    pass

# student.py
with open("college/student.py", "w") as f:
    f.write("""
def student():
    print("\\n----- Student Details -----")
    print("Name : Rahul")
    print("Class : BCA 2nd Semester")
    print("Roll No : 101")
""")

# teacher.py
with open("college/teacher.py", "w") as f:
    f.write("""
def teacher():
    print("\\n----- Teacher Details -----")
    print("Name : Sharma Sir")
    print("Subject : Python Programming")
""")

# fees.py
with open("college/fees.py", "w") as f:
    f.write("""
def fees():
    print("\\n----- Fees Details -----")
    print("Fees Status : Paid")
    print("Amount : ₹25,000")
""")
```

```python
from college.student import student
from college.teacher import teacher
from college.fees import fees

while True:
    print("\n===== COLLEGE MANAGEMENT SYSTEM =====")
    print("1. Student Details")
    print("2. Teacher Details")
    print("3. Fees Details")
    print("4. Exit")

    try:
        choice = int(input("Enter your choice: "))

        if choice == 1:
            student()

        elif choice == 2:
            teacher()

        elif choice == 3:
            fees()

        elif choice == 4:
            print("Thank You! Exiting Program...")
            break

        else:
            raise ValueError("Invalid Choice!")

    except ValueError as e:
        print("Error:", e)

    except Exception as e:
        print("Unexpected Error:", e)
```
