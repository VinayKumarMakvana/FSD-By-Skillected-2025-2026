```python
# real world problem
# class(design)
# object(actual mobile)
# attribute(detail)
# problem:
# har object ki detail alag chaiye

# constreuctor(__init__)

# parameterized constructor

# self

# next----> methods(behaviour)
# methods:a method is a function that xdefine the behaviour of an object
# attributes=object ki detail
# brand
# ram
# storage
# methods=for an action
# call()
# camera()
# music()
```

```python
class mobile:
  def __init__(self,brand,ram):
    self.brand=brand
    self.ram=ram
  def call(self):
    print(f" mobile {self.brand} is callingg..........")
  def camera(self):
    print(self.brand,"taking  picture")
  def music(self):
    print(self.brand,"playing music")
```

```python
m1=mobile("apple","8 gb")
m2=mobile("samsung","12 gb")
m3=mobile("oneplus","16 gb")
```

```python
m2.call()#method call
```
