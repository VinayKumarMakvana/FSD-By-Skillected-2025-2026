```python
# Polymorephism(Method overwiting) :- Polymorphism via Method Overriding is a runtime mechanism where a child class provides
#                                     a specific implementation of a method that is already defined in its parent class.

class Animal():
  def sound(self):
    print("animal make sound")
```

```python
# Polymorephism(Method overwiting)
class Dog(Animal):
  def sound(self):
    print("dog barks")
```

```python
d=Dog()
d.sound()
```

```python
# method overwriting
class Cat(Animal):
  def sound(delf):
    print("meow meow")
```

```python
class Animal:
    def intro(self):
        print("This is a generic animal.")

class Dog(Animal):
    def intro(self):
        print("This is a dog. Woof!")

# Overriding the parent class method
class Cat(Animal):
    def intro(self):
        print("This is a cat. Meow!")

# Demonstrating Runtime Polymorphism
animals = [Dog(), Cat(), Animal()]

for animal in animals:
    # Python automatically picks the correct method at runtime
    animal.intro()
```
