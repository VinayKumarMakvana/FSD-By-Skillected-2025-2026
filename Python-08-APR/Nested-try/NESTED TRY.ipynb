# **NESTED TRY**

```python
# Mested Try :- a try block inside another try block is called nested try
```

```python
try:
  print("Outer Try Start")
  try:
    num=int(input("Enter number: "))
    print(num)
  except ValueError:
    print("Inner value Error")
except:
  print("Outer value error")
```

```python
try:
  name=input("enter name: ")
  print(name)
  try:
    num=int(input("Enter number: "))
    print(num)
  except ValueError:
    print("Inner value Error")
except NameError:
  print("Outer NAME error")
```

```python
try:
  file=open("student.txt","w")
  file.write("Akash\n")
  file.write("Shreya\n")
  file.write("Vinay\n")
  file.close()
  print("File written successfully")
  try:
    file=open("student.txt","r")
    data=file.read()
    print(data)
    file.close()
  except FileNotFoundError:
    print("FIle note found")
except:
  print("something went wrong")
```

```python
# rise keyword :-

age=int(input("Enter your age: "))
if age<18:
  raise ValueError("You are not eligible to vote")
print("Eligible to vote")
```

```python
try:
  age=int(input("Enter your age: "))
  if age<18:
    raise ValueError("You are not eligible to vote")
  print("Eligible to vote")
except ValueError as e:
  print(e)
```

```python
try:
  password=input("password: ")
  if len(password) < 8:
    raise ValueError("password must contain at least 8 character")
  print("password accepted")
except ValueError as e:
  print(e)
len(password)
```
