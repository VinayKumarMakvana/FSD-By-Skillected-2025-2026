```python
# Q6. Create a function to check even or odd
def checknum(num):
    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

checknum(7)
```

```python
# Q7. Create a function to find greatest among 3 numbers
def greatest(a, b, c):
    if a > b and a > c:
        print(a, "is greatest")
    elif b > c:
        print(b, "is greatest")
    else:
        print(c, "is greatest")

greatest(10, 50, 30)
```

```python
# Q8. Create a function to calculate result using marks
def result(marks):
    if marks >= 90:
        print("Excellent")
    elif marks >= 70:
        print("Good")
    elif marks >= 50:
        print("Average")
    else:
        print("Fail")

result(85)
```

```python
# Q9. Create a function to count vowels in a string
def vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for a in text:
        if a in vowels:
            count += 1

    print("Total vowels =", count)

vowels("Hello Students")
```

```python
# Q10. Create a function to reverse a word
def reverse(word):
    print("Reverse =", word[::-1])

reverse("Python")
```

```python
# Q11. Create a function to print table of 7
def table():
    for i in range(1, 11):
        print("7 x", i, "=", 7 * i)

table()
```

```python
# Q15. Create a mini employee salary system using functions
def salary(emp_name, basic):
    hra = basic * 0.20
    da = basic * 0.10
    total = basic + hra + da

    print("Employee:", emp_name)
    print("Basic Salary:", basic)
    print("Total Salary:", total)

salary("Amit", 20000)
```

```python
# Q28. Create a file-based employee record system

# Features:

# Add employee
# View employees

def add_employee():
    file = open("employees.txt", "a")

    name = input("Enter employee name: ")

    file.write(name + "\n")

    file.close()

    print("Employee Added")

def view_employee():
    file = open("employees.txt", "r")

    print(file.read())

    file.close()

add_employee()
view_employee()
```
